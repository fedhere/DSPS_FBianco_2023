```python
import numpy as np
import matplotlib.pylab as plt
```

```python
np.random.seed(617)
cointoss = []
for i in range(10000):
  cointoss.append(np.round(np.random.rand()))

plt.hist(cointoss, bins=[-0.1, 0.1, 0.9, 1.1]);
#cointoss
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("H=0; T=1");
```

P(H) = 0.5


P(HH) = 0.5 * 0.5 = 0.25

P(HHT) = 0.5 * 0.5 * 0.5 = 0.125

P(2H & 1T) = P(HHT) + P(HTH) + P(THH) = 3 * 0.125  = 0.375

$P(N_H; N_T) = P(H)^{N_H} * P(T)^{N_T} * (N_H + N_T) / N_H $

```python
np.random.seed(302)
cointoss = []
for i in range(0,30):
  for i in range(10000):
    cointoss.append(np.round(np.random.rand(30)).sum())

plt.hist(cointoss, bins=np.arange(0,30));
#cointoss
plt.ylabel("frequency")
plt.xlabel("number of heads in 30 draws");
```

```python
np.random.seed(302)
cointoss = []
for i in range(0,30):
  for i in range(10000):
    cointoss.append(np.round(np.random.rand(30)).sum())

plt.hist(cointoss, bins=np.arange(0,30), density=True);
#cointoss
plt.xlabel("number of heads in 30 draws")
plt.ylabel("probability");
```

```python
print(plt.hist(cointoss, bins=np.arange(0,30), density=True)[0].sum())

plt.xlabel("number of heads in 30 draws")
plt.ylabel("probability");
```
