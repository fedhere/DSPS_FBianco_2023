```python
import numpy as np
import pylab as pl
import scipy as sp
```

```python
np.random.seed(362)
```

```python
np.random.normal?
```

```python
mean = 2
for i in range(1, 10):
  mean = mean * i
  pl.figure()
  poiss = np.random.poisson(mean, 100)
  norm = np.random.normal(loc=mean, scale=np.sqrt(mean), size=100)
  pl.hist(poiss, label="poisson")
  pl.hist(norm, label="normal", alpha=0.5)
  pl.legend()
  ks, p_ks = sp.stats.ks_2samp(poiss, norm)
  pl.title("mean = %d p-value = %.2f"%(mean, p_ks));
```

```python
pl.hist(np.random.poisson(10, 100))
pl.hist(np.random.normal(loc = 10, scale = np.sqrt(10), size=100))
```
